```python
import cv2
```

```python
import matplotlib.pyplot as plt
```

```python
Path="D:\Computer_Vision\CREDIT CARD OCR\cc5.jpeg"
img1=cv2.imread(Path)
plt.imshow(img1)
```

```python
gray_img=cv2.cvtColor(img1,cv2.COLOR_BGR2GRAY)
plt.imshow(gray_img)
```

```python
(thresh,binary_img)=cv2.threshold(gray_img,100,255,cv2.THRESH_BINARY|cv2.THRESH_OTSU)
plt.imshow(binary_img)
```

```python
import pytesseract
pytesseract.pytesseract.tesseract_cmd=r"C:\Program Files\Tesseract-OCR\tesseract.exe"
```

```python
text=pytesseract.image_to_string(binary_img, lang="eng",config="--psm 11")
print(text)
```

```python
import re
text=re.sub(r'[^0-9]+[\n\r\t]',"",text)
print(text)
```

```python
CardNumber=text[0:19]
ExpiryDate=text[19:]
Card_Number={"CardNumber":[CardNumber],"ExpiryDate":[ExpiryDate]}
Card_Number
```

```python
import pandas as pd
df=pd.DataFrame(Card_Number)
df
```
